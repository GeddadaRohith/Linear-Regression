```python
import pyttsx3
import speech_recognition as sr

engine = pyttsx3.init()

def add(a,b):
    result = a+b
    speak(f"The addition of {a} and {b} is {result}")
    
def sub(a,b):
    result = a-b
    speak(f"The subtraction of {a} and {b} is {result}")

def mul(a,b):
    result = a*b
    speak(f"The multiplication of {a} and {b} is {result}")

def div(a,b):
    result = a/b
    speak(f"The division of {a} and {b} is {result}")
    
def square(a):
    result = a**2
    speak(f"The square of {a} is {result}")
    
def cube(a):
    result = a**3
    speak(f"The cube of {a} is {result}")

def power(a,b):
    result = a**b
    speak(f"The {a} to the power of {b} is {result}")


def speak(text):
    engine.say(text)
    engine.runAndWait()
    
def listen():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        speak("yes sir, tell me")
        try:
            audio = recognizer.listen(source)
            command = recognizer.recognize_google(audio)
            return command.lower()
        except sr.UnknownValueError:
            speak("Sorry, i did not catch that")
            return ""
def main():
    speak("Enter your password")
    password = listen()
    if password == "geethika" or "rohith":
        speak("Hello rohith sir  ! iam your bot ")
        speak("I am here to help you daily purposes")
        while True:
            command = listen()
            if "hello" in command:
                speak("hello rohith sir ! how can i help you?.")
            elif "add these two numbers " in command or "unlock addition" in command:
                speak("enter the numbers sir")
                a=int(input())
                b=int(input())
                add(a,b)
            elif "subtract these two numbers" in command or "unlock subtraction" in command:
                speak("enter the numbers sir")
                a=int(input())
                b=int(input())
                sub(a,b)
            elif "multiply these two numbers" in command or "unlock multiplication" in command:
                speak("enter the numbers sir")
                a=int(input())
                b=int(input())
                mul(a,b)
            elif "division these two numbers" in command or "unlock division" in command:
                speak("enter the numbers sir")
                a=int(input())
                b=int(input())
                div(a,b)
            elif "square the number" in command or "unlock square" in command:
                speak("enter the number sir")
                a=int(input())
                square(a)
            elif "cube the numbers" in command or "unlock cube" in command:
                speak("enter the number sir")
                a=int(input())
                cube(a)
            elif "power numbers" in command or "unlock power" in command:
                speak("enter the numbers sir")
                a=int(input())
                b=int(input())
                power(a,b)
            elif "your doing good" in command or "good bot" in command:
                speak("thank you sir , your pleasure")
            elif "bye" in command or "i will talk to you later" in command:
                speak("bye sir! have a nice day")
                break
            else:
                speak("I didn't understand that . please try again" )
    else:
        speak("Worng password")
        speak("Try again")
if __name__ == "__main__":
    main()
```
