# Replace Using Mean, Median, or Mode

# Mean



```python
import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)

print(df.to_string()) 
print(x)
```

```python
import pandas as pd

df=pd.read_csv('titanic.csv')#Reading dataset

x=df["age"].mean()#finding the mean of age

df["age"].fillna(x,inplace=True)#replaceing with mean

print("mean of age is:",x)

print(df.to_string())
```

```python
import pandas as pd 
df=pd.read_csv('titanic.csv')
x=df["age"].median()
df["age"].fillna(x,inplace=True)
print("The median of age is ",x)
print(df.to_string())
```

```python
import pandas as pd

df=pd.read_csv('titanic.csv')

x=df["age"].mode()[0]

df["age"].fillna(x,inplace=True)

print(x)

print(df.to_string())
```
